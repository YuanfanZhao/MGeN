```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

def set_academic_style():
    plt.style.use('default')
    rcParams['font.family'] = 'serif'
    rcParams['font.serif'] = ['Times New Roman', 'DejaVu Serif']
    rcParams['mathtext.fontset'] = 'stix'
    rcParams['font.size'] = 12
    rcParams['axes.labelsize'] = 14
    rcParams['axes.titlesize'] = 16
    rcParams['xtick.labelsize'] = 12
    rcParams['ytick.labelsize'] = 12
    rcParams['legend.fontsize'] = 12
    rcParams['figure.figsize'] = (12, 5)
    rcParams['figure.dpi'] = 800
    rcParams['savefig.dpi'] = 800
    rcParams['axes.linewidth'] = 1.2

set_academic_style()

df = pd.read_csv('Result.csv')

KL_pivot = df.pivot(index = "True_Model", columns = "Fit_Model", values = "Rank_KL")
ISE_pivot = df.pivot(index = "True_Model", columns = "Fit_Model", values = "Rank_ISE")
MSE_pivot = df.pivot(index = "True_Model", columns = "Fit_Model", values = "Rank_MSE")

fig, axes = plt.subplots(1, 3, figsize = (18, 5))

sns.heatmap(KL_pivot, annot = True, fmt = ".4f", cmap = "YlGnBu", 
            cbar = True, ax = axes[0], linewidths = 0.5, linecolor = "white")
axes[0].set_title("KL divergence")
axes[0].set_xlabel("Fitting distribution")
axes[0].set_ylabel("Generating distribution")

sns.heatmap(ISE_pivot, annot = True, fmt = ".4f", cmap = "YlGnBu", 
            cbar = True, ax = axes[1], linewidths = 0.5, linecolor = "white")
axes[1].set_title("ISE")
axes[1].set_xlabel("Fitting distribution")
axes[1].set_ylabel("Generating distribution")

sns.heatmap(MSE_pivot, annot = True, fmt = ".4f", cmap = "YlGnBu", 
            cbar = True, ax = axes[2], linewidths = 0.5, linecolor = "white")
axes[2].set_title("MSE")
axes[2].set_xlabel("Fitting distribution")
axes[2].set_ylabel("Generating distribution")

plt.tight_layout()
plt.show()
```
