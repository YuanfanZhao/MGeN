```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

def set_academic_style():
    plt.style.use('default')
    rcParams['font.family'] = 'serif'
    rcParams['font.serif'] = ['Times New Roman', 'DejaVu Serif']
    rcParams['mathtext.fontset'] = 'stix'
    rcParams['font.size'] = 14
    rcParams['axes.labelsize'] = 16
    rcParams['axes.titlesize'] = 18
    rcParams['xtick.labelsize'] = 14
    rcParams['ytick.labelsize'] = 14
    rcParams['legend.fontsize'] = 14
    rcParams['figure.figsize'] = (12, 5)
    rcParams['figure.dpi'] = 800
    rcParams['savefig.dpi'] = 800
    rcParams['axes.linewidth'] = 1.2

set_academic_style()

df = pd.read_csv('Result.csv')

row_order = ['t', 'Laplace', 'Normal', 'Uniform']
col_order = ['NPMN', 'IGauN', 'RIGauN', 'GaN', 't', 'IGaN', 'Laplace', 'Normal']

KL_pivot = df.pivot(index = "True_Model", columns = "Fit_Model", values = "Rank_KL")
KL_pivot = KL_pivot.reindex(index = row_order, columns = col_order)
KL_text  = df.pivot(index="True_Model", columns="Fit_Model", values="KL")
KL_text = KL_text.reindex(index = row_order, columns=col_order)

ISE_pivot = df.pivot(index = "True_Model", columns = "Fit_Model", values = "Rank_ISE")
ISE_pivot = ISE_pivot.reindex(index = row_order, columns=col_order)
ISE_text  = df.pivot(index="True_Model", columns="Fit_Model", values="ISE")
ISE_text = ISE_text.reindex(index = row_order, columns=col_order)

def create_annot_matrix(val_df, rank_df):
    annot_df = pd.DataFrame(index=val_df.index, columns=val_df.columns)
    for i in range(val_df.shape[0]):
        for j in range(val_df.shape[1]):
            val = val_df.iloc[i, j]
            rank = rank_df.iloc[i, j]
            annot_df.iloc[i, j] = f"{val:.4f}\n({rank:.4f})"
    return annot_df

annot_KL = create_annot_matrix(KL_text, KL_pivot)
annot_ISE = create_annot_matrix(ISE_text, ISE_pivot)

fig, axes = plt.subplots(1, 2, figsize = (18, 5))

sns.heatmap(KL_pivot, annot = annot_KL, fmt = "", cmap = "YlGnBu", 
            cbar = True, ax = axes[0], linewidths = 0.5, linecolor = "white")
axes[0].set_title("KL divergence")
axes[0].set_xlabel("Fitting distribution")
axes[0].set_ylabel("Dominant distribution")

sns.heatmap(ISE_pivot, annot = annot_ISE, fmt = "", cmap = "YlGnBu", 
            cbar = True, ax = axes[1], linewidths = 0.5, linecolor = "white")
axes[1].set_title("ISE")
axes[1].set_xlabel("Fitting distribution")
axes[1].set_ylabel("Dominant distribution")

plt.tight_layout()
plt.savefig('plot19', dpi=800, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
